```c
//%cflags:-lpthread

#include <sys/types.h>
#include <sys/stat.h>
#include <fcntl.h>
#include <stdlib.h>
#include <stdio.h>
#include <sys/mman.h>
#include <semaphore.h>

#define MAX_BUFFER 1024 /* tamaño del buffer */
#define DATOS_A_PRODUCIR 100000 /* datos a producir */

void Productor(int* buffer, sem_t* huecos, sem_t* elementos) {
  int pos = 0; /* posicion dentro del buffer */
  int dato; /* dato a producir */
  int i;
  for(i=0; i < DATOS_A_PRODUCIR; i++ ) {
    dato = i; /* producir dato */
    printf("Esperamos el hueco %d.\n", i);
    sem_wait(huecos); /* un hueco menos */
    printf("Fin de la espera.\n");
    buffer[pos] = dato;
    pos = (pos + 1) % MAX_BUFFER;
    sem_post(elementos); /* un elemento mas */
  }
  return;
}

void main(int argc, char *argv[]) {
  // BUFFER:
  int shd = open("BUFFER", O_RDWR|O_CREAT, S_IRWXU);
  ftruncate(shd, MAX_BUFFER * sizeof(int));
  int* buffer = mmap(0, MAX_BUFFER*sizeof(int), PROT_READ|PROT_WRITE, MAP_SHARED, shd, 0);
  close(shd);

  // SEMAFOROS
  sem_t* elementos = sem_open("ELEMENTOS", O_CREAT, 0644, 0);
  sem_t* huecos = sem_open("HUECOS", O_CREAT, 0644, MAX_BUFFER);

  Productor(buffer, huecos, elementos);

  // Libero buffer
  munmap(buffer, MAX_BUFFER * sizeof(int));
  unlink("BUFFER"); /* borrar el objeto de memoria */
  // Libero semáforos
  sem_close(elementos);
  sem_close(huecos);
  sem_unlink("ELEMENTOS");
  sem_unlink("HUECOS");
}
```

```c
//%cflags:-lpthread

#include <sys/types.h>
#include <sys/stat.h>
#include <fcntl.h>
#include <stdlib.h>
#include <stdio.h>
#include <sys/mman.h>
#include <semaphore.h>

#define MAX_BUFFER 1024
#define DATOS_A_PRODUCIR 100000

void Consumidor(int* buffer, sem_t* huecos, sem_t* elementos) {
  int pos = 0;
  int i, dato;
  for(i=0; i < DATOS_A_PRODUCIR; i++ ) {
    printf("Esperamos un elemento.\n");
    sem_wait(elementos); /* un elemento menos */
    printf("Fin de la espera.\n");
    dato = buffer[pos];
    pos = (pos + 1) % MAX_BUFFER;
    sem_post(huecos);    /* un hueco mas */
    printf("Consume %d \n", dato); /* cosumir dato */
  }
  return;
}


void main(int argc, char *argv[]) {
  int shd = open("BUFFER", O_RDONLY);
  int* buffer = (int *) mmap(0, MAX_BUFFER * sizeof(int), PROT_READ, MAP_SHARED, shd, 0);

  sem_t* elementos = sem_open("ELEMENTOS", 0);
  sem_t* huecos = sem_open("HUECOS", 0);
  Consumidor(buffer, huecos, elementos);
  
  munmap(buffer, MAX_BUFFER * sizeof(int));
  close(shd);
  sem_close(elementos);
  sem_close(huecos);
}
```
