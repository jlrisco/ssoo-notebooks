En este ejemplo vamos a ver cómo se puede implementar el problema de los Lectores/Escritores con mutex.

Tenemos dos cerrojos: El primero es un cerrojo global para bloquear la escritura de dato. El segundo es un cerrojo sobre la variable que cuenta el número de lectores.

```c
//%cflags:-lpthread

#include <stdio.h>
#include <stdlib.h>
#include <unistd.h>
#include <pthread.h>

int dato = 5;           /* recurso */
int n_lect = 0;         /* numero de lectores*/
pthread_mutex_t mutex;  /*controlar el acceso a dato*/
pthread_mutex_t m_lect; /*controla la variable n_lect*/

void* Lector(void* ptr);
void* Escritor(void* ptr);

void main(int argc, char *argv[]) {
  pthread_t th1, th2, th3, th4;
  pthread_mutex_init(&mutex, NULL);
  pthread_mutex_init(&m_lect, NULL);
  pthread_create(&th1, NULL, Lector, NULL);
  pthread_create(&th2, NULL, Escritor, NULL);
  pthread_create(&th3, NULL, Lector, NULL);
  pthread_create(&th4, NULL, Escritor, NULL);
  pthread_join(th1, NULL);
  pthread_join(th2, NULL);
  pthread_join(th3, NULL);
  pthread_join(th4, NULL);
  pthread_mutex_destroy(&mutex);
  pthread_mutex_destroy(&m_lect);
  exit(0);
}

/*codigo del lector */
void* Lector(void* ptr) {
  while(dato<50) { // <- INCORRECTO, PARA QUE SEA FINITO
    pthread_mutex_lock(&m_lect);
    n_lect ++;
    if (n_lect == 1) {
      pthread_mutex_lock(&mutex);
    }
    pthread_mutex_unlock(&m_lect);
    printf("%d\n", dato); /*leer*/
    fflush(stdout);
    pthread_mutex_lock(&m_lect);
    n_lect--;
    if (n_lect == 0) {
      pthread_mutex_unlock(&mutex);
    }
    pthread_mutex_unlock(&m_lect);
    sleep(1);
  }
}

/*codigo del escritor */
void* Escritor(void* ptr) {
  while(dato<50) { // <- INCORRECTO, PARA QUE SEA FINITO
    pthread_mutex_lock(&mutex);
    dato = dato + 2; /*modificar el recurso */
    pthread_mutex_unlock(&mutex);
    sleep(1);
  }
}
```

Ahora vamos a ver si una hebra puede abrir un cerrojo echado por otra hebra:

```c
//%cflags:-lpthread

#include <stdio.h>
#include <stdlib.h>
#include <pthread.h>

pthread_mutex_t cerrojo;  

void* Hebra1(void* ptr);
void* Hebra2(void* ptr);

void main(int argc, char *argv[]) {
  pthread_t th1, th2;
  pthread_mutex_init(&cerrojo, NULL);
  pthread_create(&th1, NULL, Hebra1, NULL);
  pthread_create(&th2, NULL, Hebra2, NULL);
  pthread_join(th1, NULL);
  pthread_join(th2, NULL);
  pthread_mutex_destroy(&cerrojo);
  exit(0);
}

/*codigo de la hebra */
void* Hebra1(void* ptr) {
  int i;
  printf("H1: Hola, soy la hebra1. Voy a echar el cerrojo.\n");
  pthread_mutex_lock(&cerrojo);
  printf("H1: ... ya está. Se supone que nadie debe pasar el lock.\n");
  printf("H1: ... me meto en un bucle un poco largo.\n");
  for(i=0; i<1000000; i++) {
  }
  printf("H1: ... y ahora abro el cerrojo.\n");
  pthread_mutex_unlock(&cerrojo);
  printf("H1: ... ya está.\n");
}

/*codigo de la hebra 2 */
void* Hebra2(void* ptr) {
  int i;
  printf("H2: Hola, soy la hebra2. Me meto en un bucle corto.\n");
  for(i=0; i<10000; i++) {
  }
  printf("H2: ... y ahora abro el cerrojo.\n");
  pthread_mutex_unlock(&cerrojo);
  printf("H2: ... lo vuelvo a cerrar para ver si puedo pasar.\n");
  pthread_mutex_lock(&cerrojo);
  printf("H2: ... y me meto en un bucle un poco largo.\n");
  for(i=0; i<1000000; i++) {
  }
  printf("H2: ... ya está.\n");
}
```
